```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of columns related to energy
energy_columns = ['Electricity_Quantity', 'NaturalGas_Quantity', 'FuelOil12_Quantity',
                   'FuelOil46_Quantity', 'Propane_Quantity', 'Coal_Quantity',
                   'Wood_Quantity', 'DistrictHeating_Quantity', 'DistrictCooling_Quantity']

# Subset DataFrame to include only energy-related columns
energy_df = df[energy_columns]

# Plot the distribution of each energy-related column
plt.figure(figsize=(12, 8))
for column in energy_df.columns:
    sns.histplot(energy_df[column], kde=True, label=column)

plt.title('Distribution of Energy Usage')
plt.xlabel('Energy Quantity')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
correlation_matrix = df[['Electricity_Quantity', 'NaturalGas_Quantity', 'FuelOil12_Quantity', 'Propane_Quantity', 'Coal_Quantity', 'Wood_Quantity', 'DistrictHeating_Quantity', 'DistrictCooling_Quantity']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix: Energy Consumption')
plt.show()
```

```python
sns.pairplot(df[['Electricity_Quantity', 'NaturalGas_Quantity', 'FuelOil12_Quantity', 'GHG Emissions KG']])
plt.suptitle('Pairplot: Energy Sources and GHG Emissions', y=1.02)
plt.show()
```

```python
#OUTLIER DETECTION AND HANDLING
# Example: Boxplot to identify outliers in GHG emissions
sns.boxplot(x='Sector', y='GHG Emissions KG', data=df)
plt.title('Boxplot: GHG Emissions by Sector')
plt.xlabel('Sector')
plt.ylabel('GHG Emissions (KG)')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import numpy as np

# Example: Logarithmic transformation of GHG emissions
df['Log_GHG_Emissions'] = np.log1p(df['GHG Emissions KG'])
```

```python
pivot_df = df.pivot_table(values='Electricity_Quantity', index='Sector', columns='SubSector', aggfunc='sum')
sns.heatmap(pivot_df, cmap='viridis', annot=True, fmt='.0f')
plt.title('Energy Consumption by Sector and Subsector')
```
